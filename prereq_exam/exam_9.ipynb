### Q9)
Using matplotlib plot the following functions on the domain: x = 0.5 to 100.0 in steps of 0.5.

### Answer)

1)
\begin{equation}
x = y
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the domain
x_values = np.arange(0.5, 100.5, 0.5)
```

```python
# Define the functions
y_values = x_values

# Plot the functions
plt.figure(figsize=(8, 6))

# Plot x = y
plt.plot(x_values, y_values, label='x = y', color='blue')

# Set labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of x = y')

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

2)
\begin{equation}
y = x^2
\end{equation}


```python
# Define the functions
y_values = x_values**2

plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='y = x^2', color='blue')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = x^2')
plt.legend()
plt.grid(True)
plt.show()
```

3)
\begin{equation}
y = \frac{x^3}{100}
\end{equation}

```python
y_values = x_values**3/100

plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='y = x^3/100', color='blue')
plt.xlabel('x')
plt.ylabel('y')

plt.title('Plot of y = x^3/100')
plt.legend()
plt.grid(True)

plt.show()
```

4)
\begin{equation}
y = sin(x)
\end{equation}

```python
y_values = np.sin(x_values)

plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='y = sin(x)', color='blue')
plt.xlabel('x')
plt.ylabel('y')

plt.title('Plot of y = sin(x)')
plt.legend()
plt.grid(True)

plt.show()
```

5)
\begin{equation}
y = \frac{sin(x)}{x}
\end{equation}

```python
y_values = np.sin(x_values)/x_values

plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='y = sin(x)/x', color='blue')
plt.xlabel('x')
plt.ylabel('y')

plt.title('Plot of y = sin(x)/x')
plt.legend()
plt.grid(True)

plt.show()
```

6)
\begin{equation}
y = log(x)
\end{equation}

```python
y_values = np.log(x_values)

plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='y = log(x)', color='blue')
plt.xlabel('x')
plt.ylabel('y')

plt.title('Plot of y = log(x)')
plt.legend()
plt.grid(True)

plt.show()
```

7)
\begin{equation}
y = e^x
\end{equation}

```python
y_values = np.exp(x_values)

plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='y = e^x', color='blue')
plt.xlabel('x')
plt.ylabel('y')

plt.title('Plot of y = e^x')
plt.legend()
plt.grid(True)

plt.show()
```
